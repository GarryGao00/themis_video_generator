```python
import cv2
import matplotlib.pyplot as plt
import sys; sys.path.append('../')
import video_generator as vg
```

### Original images downloaded from UCalgary

```python
# read in example images
path1 = 'gako20161013123527.pgm'
path2 = 'rank20200104000000.pgm'
img1 = cv2.imread(path1, 0)
img2 = cv2.imread(path2, 0)

# add image labels
cv2.putText(img1, 'img1', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
cv2.putText(img2, 'img2', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)

# Horizontally concatenate the images and display
result = cv2.hconcat([img1, img2])
fig, ax = plt.subplots()
im = ax.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

## Different image processing techniques included

### Relu

```python
img1_relu = vg._relu(path1, low = 0, pivot=0.02, ratio=1.7)
img2_relu = vg._relu(path2, low = 0, pivot=0.02, ratio=1.7)
# add image labels
cv2.putText(img1_relu, 'img1_relu', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
cv2.putText(img2_relu, 'img2_relu', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
# Horizontally concatenate the images and display
result = cv2.hconcat([img1_relu, img2_relu])
fig, ax = plt.subplots()
im = ax.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

### Bytescale

```python
img1_bytescale = vg._bytescale(path1)
img2_bytescale = vg._bytescale(path2)
# add image labels
cv2.putText(img1_bytescale, 'img1_bytescale', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
cv2.putText(img2_bytescale, 'img2_bytescale', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
# Horizontally concatenate the images and display
result = cv2.hconcat([img1_bytescale, img2_bytescale])
fig, ax = plt.subplots()
im = ax.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

### Eqhist

```python
img1_eq = vg._eqhist(path1)
img2_eq = vg._eqhist(path2)
# add image labels
cv2.putText(img1_eq, 'img1_eq', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
cv2.putText(img2_eq, 'img2_eq', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
# Horizontally concatenate the images and display
result = cv2.hconcat([img1_eq, img2_eq])
fig, ax = plt.subplots()
im = ax.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

### Clahe

```python
img1_clahe = vg._clahe(path1)
img2_clahe = vg._clahe(path2)
# add image labels
cv2.putText(img1_clahe, 'img1_clahe', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
cv2.putText(img2_clahe, 'img2_clahe', (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, 255)
# Horizontally concatenate the images and display
result = cv2.hconcat([img1_clahe, img2_clahe])
fig, ax = plt.subplots()
im = ax.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

### Stitched together

```python
def concat_tile(im_list_2d):
    return cv2.vconcat([cv2.hconcat(im_list_h) for im_list_h in im_list_2d])
im_tile = concat_tile([[img1, img1_relu, img1_bytescale, img1_eq, img1_clahe],
                       [img2, img2_relu, img2_bytescale, img2_eq, img2_clahe]])
fig, ax = plt.subplots(figsize=(15,6))
im = ax.imshow(im_tile, cmap='gray')
plt.axis('off')
plt.show()
```
